```python
import dask.dataframe as dd
import pandas as pd
```

```python
pdf = pd.DataFrame({
    "id": range(1_000_000),
    "category": [i % 5 for i in range(1_000_000)]
})

df = dd.from_pandas(pdf, npartitions=10)
```

```python
df.shape[0].compute()
```

```python
df.groupby("category").size().compute()
```

## Insights from Big Data Analysis using Dask

- Successfully processed 1,000,000 records using Dask
- Performed distributed group-by operations efficiently
- Demonstrates scalability and big data handling capability
- Dask provides an efficient alternative to Spark on Windows
